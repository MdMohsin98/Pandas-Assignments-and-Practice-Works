```python
#By Dictionary
weather_data = {'day':['1/1/2017','1/2/2017','1/3/2017'],'temperature':[32,35,28],'windspeed':[6,7,2],'event':['Rain','Sunny','Snow']}
df = pd.DataFrame(weather_data)
df
```

```python
#By using tuples
weather_data = [('1/1/2017',32,6,'Rain'),('1/2/2017',35,7,'Sunny'),('1/3/2017',28,2,'Snow')]
df = pd.DataFrame(weather_data,columns = ["day","temperature","windspeed","event"])
df
```

```python
#By using dictionary second way
weather_data = [{'day':'1/1/2017','temp':32, 'windspeed':6,'event':'rain'},
                {'day':'1/2/2017','temp':35, 'windspeed':7,'event':'sunny'},
                {'day':'1/3/2017','temp':28, 'windspeed':2,'event':'snow'}]
df = pd.DataFrame(weather_data)
df
```
