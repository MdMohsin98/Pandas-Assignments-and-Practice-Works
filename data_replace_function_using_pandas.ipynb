```python
# Data Replace Functions
import pandas as pd
import numpy as np
df = pd.read_csv("C:\\Users\\Dell\\PycharmProjects\\coding\\venv\\excel1.csv")
df
```

```python
# Replacing an irrelevent data to NaN value
new_df = df.replace(-99999, np.NaN)
new_df
```

```python
# Replacing more than one irrelevent data to NaN value
new_df = df.replace([-99999, -88888], np.NaN)
new_df
```

```python
# Replacing an irrelevent data to NaN value
new_df = df.replace([-99999, -88888, 0], np.NaN)
new_df
```

```python
new_df = df.replace({'serial' : [-99999, -88888],
                    'roll' : [-99999, -88888],
                     'math' : [-99999, -88888],
                     'total' : [-99999,-88888,0]
                    }, np.NaN)
new_df
```

```python
#Detecting Patterns : Regex Functions
new_df = df.replace({
    'phy' : '[A-Za-z]',
    'che' : '[A-Za-z]'},'',regex = True)
new_df
```
